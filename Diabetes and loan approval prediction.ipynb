```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import scipy.stats as stats

# Functie pentru calcularea lui Cramer's V
def cramers_v(x, y):
    confusion_matrix = pd.crosstab(x, y)
    chi2 = stats.chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2 / n
    r, k = confusion_matrix.shape
    phi2corr = max(0, phi2 - ((k-1)*(r-1))/(n-1))
    rcorr = r - ((r-1)**2)/(n-1)
    kcorr = k - ((k-1)**2)/(n-1)
    return np.sqrt(phi2corr / min((kcorr-1), (rcorr-1)))

# Functie pentru calcularea matricei Cramer's V
def calculate_cramers_v_matrix(df, cols):
    matrix = np.zeros((len(cols), len(cols)))
    for i in range(len(cols)):
        for j in range(len(cols)):
            matrix[i, j] = cramers_v(df[cols[i]], df[cols[j]])
    return pd.DataFrame(matrix, index=cols, columns=cols)

# Atributele categorice
Diabet_categoric = ['HealthcareInterest', 'PreCVA', 'RoutineChecks', 'CompletedEduLvl', 'alcoholAbuse', 'cholesterol_ver', 'vegetables', 'HighBP', 'Unprocessed_fructose', 'Jogging', 'IncreasedChol', 'gender', 'HealthScore', 'myocardial_infarction', 'SalaryBraket', 'Cardio', 'ImprovedAveragePulmonaryCapacity', 'Smoker', 'Diabetes']
Credit_Risk_categoric = ['residential_status', 'loan_purpose', 'loan_rating', 'credit_history_default_status', 'stability_rating', 'loan_approval_status']


# Calcularea matricei Cramer's V pentru DiabetTrain
cramers_v_matrix_diabet = calculate_cramers_v_matrix(csv_Diabet_train, Diabet_categoric)

# Vizualizarea matricei Cramer's V pentru Diabet Train
plt.figure(figsize=(14, 12))
sns.heatmap(cramers_v_matrix_diabet, annot=True, cmap='coolwarm', vmin=0, vmax=1)
plt.title('Matricea Cramer\'s V pentru Atributele Categorice din Diabet Train')
plt.show()



# Calcularea matricei Cramer's V pentru Diabet Test
cramers_v_matrix_diabet = calculate_cramers_v_matrix(csv_Diabet_test, Diabet_categoric)

# Vizualizarea matricei Cramer's V pentru Diabet Test
plt.figure(figsize=(14, 12))
sns.heatmap(cramers_v_matrix_diabet, annot=True, cmap='coolwarm', vmin=0, vmax=1)
plt.title('Matricea Cramer\'s V pentru Atributele Categorice din Diabet Test')
plt.show()



# Calcularea matricei Cramer's V pentru Credit_Risk Train
cramers_v_matrix_credit_risk = calculate_cramers_v_matrix(csv_Credit_Risk_train, Credit_Risk_categoric)

# Vizualizarea matricei Cramer's V pentru Credit_Risk Train
plt.figure(figsize=(14, 12))
sns.heatmap(cramers_v_matrix_credit_risk, annot=True, cmap='coolwarm', vmin=0, vmax=1)
plt.title('Matricea Cramer\'s V pentru Atributele Categorice din Credit_Risk Train')
plt.show()



# Calcularea matricei Cramer's V pentru Credit_Risk Test
cramers_v_matrix_credit_risk = calculate_cramers_v_matrix(csv_Credit_Risk_test, Credit_Risk_categoric)

# Vizualizarea matricei Cramer's V pentru Credit_Risk Test
plt.figure(figsize=(14, 12))
sns.heatmap(cramers_v_matrix_credit_risk, annot=True, cmap='coolwarm', vmin=0, vmax=1)
plt.title('Matricea Cramer\'s V pentru Atributele Categorice din Credit_Risk Test')
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer

# Incarcarea datelor
csv_Diabet_train = pd.read_csv(cale1)
csv_Diabet_test = pd.read_csv(cale2)
csv_Credit_Risk_train = pd.read_csv(cale3)
csv_Credit_Risk_test = pd.read_csv(cale4)

# Functie pentru detectarea si inlocuirea valorilor extreme
def replace_extreme_values(df, numeric_cols):
    for col in numeric_cols:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Inlocuirea valorilor extreme cu NaN
        df[col] = np.where((df[col] < lower_bound) | (df[col] > upper_bound), np.nan, df[col])

    return df

# Aplicarea functiei pentru identificarea si inlocuirea valorilor extreme in seturile de date
csv_Diabet_train = replace_extreme_values(csv_Diabet_train, Diabet_numeric)
csv_Diabet_test = replace_extreme_values(csv_Diabet_test, Diabet_numeric)
csv_Credit_Risk_train = replace_extreme_values(csv_Credit_Risk_train, Credit_Risk_numeric)
csv_Credit_Risk_test = replace_extreme_values(csv_Credit_Risk_test, Credit_Risk_numeric)

# Crearea imputatorului pentru a inlocui valorile NaN (care acum includ si valorile extreme inlocuite)
imputer_numeric = SimpleImputer(strategy='mean')

# Aplicarea imputarii pentru setul de date Diabet Train
csv_Diabet_train[Diabet_numeric] = imputer_numeric.fit_transform(csv_Diabet_train[Diabet_numeric])

# Aplicarea imputarii pentru setul de date Diabet Test
csv_Diabet_test[Diabet_numeric] = imputer_numeric.fit_transform(csv_Diabet_test[Diabet_numeric])

# Aplicarea imputarii pentru setul de date Credit Risk Train
csv_Credit_Risk_train[Credit_Risk_numeric] = imputer_numeric.fit_transform(csv_Credit_Risk_train[Credit_Risk_numeric])

# Aplicarea imputarii pentru setul de date Credit Risk Test
csv_Credit_Risk_test[Credit_Risk_numeric] = imputer_numeric.fit_transform(csv_Credit_Risk_test[Credit_Risk_numeric])

# Verificarea daca mai sunt date lipsa dupa imputare
print("\nAtribute cu date lipsa in Diabet Train după imputarea valorilor extreme:")
print(csv_Diabet_train.isnull().sum())

print("\nAtribute cu date lipsa in Diabet Test după imputarea valorilor extreme:")
print(csv_Diabet_test.isnull().sum())

print("\nAtribute cu date lipsa in Credit Risk Train dupa imputarea valorilor extreme:")
print(csv_Credit_Risk_train.isnull().sum())

print("\nAtribute cu date lipsa in Credit Risk Test dupa imputarea valorilor extreme:")
print(csv_Credit_Risk_test.isnull().sum())
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats

# Incarcarea datelor
csv_Diabet_train = pd.read_csv(cale1)
csv_Diabet_test = pd.read_csv(cale2)
csv_Credit_Risk_train = pd.read_csv(cale3)
csv_Credit_Risk_test = pd.read_csv(cale4)

# Functie pentru calcularea lui Cramer's V
def cramers_v(x, y):
    confusion_matrix = pd.crosstab(x, y)
    chi2 = stats.chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2 / n
    r, k = confusion_matrix.shape
    phi2corr = max(0, phi2 - ((k-1)*(r-1))/(n-1))
    rcorr = r - ((r-1)**2)/(n-1)
    kcorr = k - ((k-1)**2)/(n-1)
    return np.sqrt(phi2corr / min((kcorr-1), (rcorr-1)))

# Functie pentru calcularea matricei Cramer's V
def calculate_cramers_v_matrix(df, cols):
    matrix = np.zeros((len(cols), len(cols)))
    for i in range(len(cols)):
        for j in range(len(cols)):
            matrix[i, j] = cramers_v(df[cols[i]], df[cols[j]])
    return pd.DataFrame(matrix, index=cols, columns=cols)

# Pragul de corelatie
threshold = 0.9

# Identificarea si eliminarea atributelor numerice puternic corelate in Diabet_train
correlation_matrix_diabet = csv_Diabet_train[Diabet_numeric].corr().abs()
upper_tri_diabet = correlation_matrix_diabet.where(np.triu(np.ones(correlation_matrix_diabet.shape), k=1).astype(bool))
to_drop_diabet_numeric = [column for column in upper_tri_diabet.columns if any(upper_tri_diabet[column] > threshold)]

# Identificarea si eliminarea atributelor categorice puternic corelate in Diabet_train
cramers_v_matrix_diabet = calculate_cramers_v_matrix(csv_Diabet_train, Diabet_categoric)
upper_tri_cramers_diabet = cramers_v_matrix_diabet.where(np.triu(np.ones(cramers_v_matrix_diabet.shape), k=1).astype(bool))
to_drop_diabet_categoric = [column for column in upper_tri_cramers_diabet.columns if any(upper_tri_cramers_diabet[column] > threshold)]

# Eliminarea atributelor puternic corelate in Diabet_train si Diabet_test
csv_Diabet_train.drop(columns=to_drop_diabet_numeric + to_drop_diabet_categoric, inplace=True)
csv_Diabet_test.drop(columns=to_drop_diabet_numeric + to_drop_diabet_categoric, inplace=True)

# Identificarea si eliminarea atributelor numerice puternic corelate in Credit_Risk_train
correlation_matrix_credit_risk = csv_Credit_Risk_train[Credit_Risk_numeric].corr().abs()
upper_tri_credit_risk = correlation_matrix_credit_risk.where(np.triu(np.ones(correlation_matrix_credit_risk.shape), k=1).astype(bool))
to_drop_credit_risk_numeric = [column for column in upper_tri_credit_risk.columns if any(upper_tri_credit_risk[column] > threshold)]

# Identificarea si eliminarea atributelor categorice puternic corelate in Credit_Risk_train
cramers_v_matrix_credit_risk = calculate_cramers_v_matrix(csv_Credit_Risk_train, Credit_Risk_categoric)
upper_tri_cramers_credit_risk = cramers_v_matrix_credit_risk.where(np.triu(np.ones(cramers_v_matrix_credit_risk.shape), k=1).astype(bool))
to_drop_credit_risk_categoric = [column for column in upper_tri_cramers_credit_risk.columns if any(upper_tri_cramers_credit_risk[column] > threshold)]

# Eliminarea atributelor puternic corelate in Credit_Risk_train
csv_Credit_Risk_train.drop(columns=to_drop_credit_risk_numeric + to_drop_credit_risk_categoric, inplace=True)
csv_Credit_Risk_test.drop(columns=to_drop_credit_risk_numeric + to_drop_credit_risk_categoric, inplace=True)

# Afisarea atributelor eliminate
print("Atribute numerice eliminate din Diabet_train:", to_drop_diabet_numeric)
print("Atribute categorice eliminate din Diabet_train:", to_drop_diabet_categoric)
print("Atribute numerice eliminate din Credit_Risk_train:", to_drop_credit_risk_numeric)
print("Atribute categorice eliminate din Credit_Risk_train:", to_drop_credit_risk_categoric)

# Afisarea seturilor de date finale
print("\nSetul de date Diabet_train dupa eliminarea atributelor puternic corelate:")
print(csv_Diabet_train.head())

print("\nSetul de date Diabet_test dupa eliminarea atributelor puternic corelate:")
print(csv_Diabet_test.head())

print("\nSetul de date Credit_Risk_train dupa eliminarea atributelor puternic corelate:")
print(csv_Credit_Risk_train.head())

print("\nSetul de date Credit_Risk_test dupa eliminarea atributelor puternic corelate:")
print(csv_Credit_Risk_train.head())
```
